```python
#tokenize documents

text_tokens_X = word_tokenize(X)
text_tokens_Y = word_tokenize(Y)

# remove stopwords
tokens_without_sw_X = [word for word in text_tokens_X if not word in stopwords.words()]
tokens_without_sw_Y = [word for word in text_tokens_Y if not word in stopwords.words()]
```

```python
# now that the stopwords have been removed, put them back in a string format

filtered_X = (" ").join(tokens_without_sw_X)
filtered_Y = (" ").join(tokens_without_sw_Y)

# print(filtered_Y)
```

```python
# define the corpus
corpus_xy = [filtered_X, filtered_Y]
```

```python
# define the tfidf vectorizer
tfidf_vectorizer = TfidfVectorizer()
```

```python
# create a tfidf vector
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus_xy)
```

```python
# obtain the cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
print(cosine_sim)
```
